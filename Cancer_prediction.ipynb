```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Data Loading
data = pd.read_csv('data-cancer.csv')
data
```

```python
# Dataset Dimensions
data.shape
```

```python
#Label Encoding: Diagnosis Column
data['diagnosis']=data['diagnosis'].map({'M':1,'B':0})
data.head()
```

```python
#Diagnosis Distribution Visualization. Plotted the distribution of diagnosis labels (Benign vs. Malignant) to understand the class balance in the dataset:
A = data['diagnosis'].value_counts()
sns.barplot(x=A.index, y=A.values)
plt.title("Distribution of Diagnosis")
plt.xlabel("Diagnosis")
plt.ylabel("Count")
plt.show()
```

```python
# Data Cleaning
# Dropping Irrelevant Features: 
del data['id']
data.head()

# Missing Value Check - 
data.isnull().sum()
```

```python
# Splitting Features and Target
# We now split the dataset into independent features X and the target variable y:
x = data.loc[:,data.columns[1:]]
y = data['diagnosis']
```

```python
#Train-Test Split divide the data into training and testing sets using an 80-20 split:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
#Model Training and Prediction: Decision Tree Classifier 
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train) 
y_pred = dt.predict(X_test)
y_pred
```

```python
#Evaluating model performance 
from sklearn.metrics import accuracy_score
y_pred = dt.predict(X_test) 
acc = accuracy_score(y_test, y_pred)
print(acc)
```

```python
#Tuning Decision Tree Depth To find the optimal tree depth for our Decision Tree Classifier,evaluated training accuracy 
# and 10-fold cross-validation scores across multiple depths:

from sklearn.model_selection import cross_val_score 
depths=range(1,21)
train_scores = []
val_scores = []
for depth in depths:
  dt = DecisionTreeClassifier(max_depth=depth) 
  dt.fit(X_train, y_train) 
  trainAccuracy = accuracy_score(y_train, dt.predict(X_train)) 
  dt = DecisionTreeClassifier(max_depth=depth) 
  valAccuracy = cross_val_score(dt, X_train, y_train, cv=10)
  train_scores.append(trainAccuracy)
  val_scores.append(valAccuracy)
  print("Depth  : ", depth, " Training Accuracy : ", trainAccuracy, " Cross val score : " ,np.mean(valAccuracy))
```

```python
import matplotlib.pyplot as plt

# Plotting
plt.figure(figsize=(10,6))
plt.plot(depths, train_scores, label='Training Accuracy', marker='o')
plt.plot(depths, val_scores, label='Cross-Validation Accuracy', marker='s')
plt.xlabel("Tree Depth")
plt.ylabel("Accuracy")
plt.title("Accuracy vs. Tree Depth")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
Insight:
The Decision Tree model effectively classified tumors as benign or malignant.
Optimal depth selection helped prevent overfitting and improved generalization.
Key features like radius_mean and concavity_mean played a major role in predictions.
Achieved high accuracy with a simple, interpretable model suitable for medical use.
```
